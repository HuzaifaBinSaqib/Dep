```python
df = df.drop(['country','gender'], axis = 1)
df_churned = df[df['churn'] == 1]
df_churned.head(10)
```

```python
y = df_churned['churn']
x = df_churned.copy()
x = x.drop(['balance'], axis = 1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify = y)
```

```python
decision_tree = DecisionTreeClassifier(random_state = 42)
decision_tree.fit(x_train, y_train)
```

```python
decision_tree = DecisionTreeClassifier(random_state = 42)
decision_tree.fit(x_train, y_train)
pred = decision_tree.predict(x_test)

print("Accuracy:" , "%.3f"% accuracy_score(y_test, pred))
print("Recall Score:" , "%.3f"% recall_score(y_test, pred))
print("Precision Score:" , "%.3f"% precision_score(y_test, pred))
print("F1 Score:" , "%.3f"% f1_score(y_test, pred))
```
